```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
movies.head()
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
credits.head()
```

```python
movies = movies.drop(["budget","homepage","original_language","popularity","production_companies","keywords","overview","production_countries","release_date","revenue","runtime","spoken_languages","status","original_title"], axis=1)
```

```python
credits = credits.drop(["cast","crew"], axis =1)
```

```python
movies.rename(columns={'id':'movie_id'}, inplace= True)
movies.head()
```

```python
DF = movies.merge(credits, on= ["movie_id","title"])
```

```python
DF = DF.drop(["genres","vote_count"], axis=1)
```

```python
DF.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(24,6))
ax = sns.countplot(x="vote_average", data= DF)
```
